```python
import os
import json
```

```python
def load_data(file_name):
    data=[]
    with open(file_name) as f:
        while True:
            line=f.readline()
            if not line:
                break
    #        print(line)
            data.append(json.loads(line))
#    print(data)
    return data

data=load_data("log/2019-07-12_02-56-26/navdata.txt")
for d in data:
    print(d,type(d))
```
